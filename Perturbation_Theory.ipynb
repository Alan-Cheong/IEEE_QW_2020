```python
import numpy as np
import matplotlib.pyplot as plt

# Set up the grid
L = 1.0  # Length of the square well
N = 100  # Number of points in each direction
x = np.linspace(0, L, N)
y = np.linspace(0, L, N)
X, Y = np.meshgrid(x, y)

# Define the wavefunctions
psi_21 = 2 * np.sin(2 * np.pi * X / L) * np.sin(np.pi * Y / L)
psi_12 = 2 * np.sin(np.pi * X / L) * np.sin(2 * np.pi * Y / L)
psi_1_prime = (1 / np.sqrt(2)) * (psi_21 + psi_12)
psi_2_prime = (1 / np.sqrt(2)) * (psi_21 - psi_12)

# Compute probability densities
prob_21 = np.abs(psi_21)**2
prob_12 = np.abs(psi_12)**2
prob_1_prime = np.abs(psi_1_prime)**2
prob_2_prime = np.abs(psi_2_prime)**2

# Create the plots
fig, axes = plt.subplots(2, 2, figsize=(10, 10))

# Plot psi_2,1
im1 = axes[0, 0].imshow(prob_21, origin='lower', extent=[0, L, 0, L], cmap='viridis')
axes[0, 0].set_title(r'$|\psi_{2,1}|^2$')
axes[0, 0].set_xlabel('x')
axes[0, 0].set_ylabel('y')
plt.colorbar(im1, ax=axes[0, 0])

# Plot psi_1,2
im2 = axes[0, 1].imshow(prob_12, origin='lower', extent=[0, L, 0, L], cmap='viridis')
axes[0, 1].set_title(r'$|\psi_{1,2}|^2$')
axes[0, 1].set_xlabel('x')
axes[0, 1].set_ylabel('y')
plt.colorbar(im2, ax=axes[0, 1])

# Plot psi_1_prime
im3 = axes[1, 0].imshow(prob_1_prime, origin='lower', extent=[0, L, 0, L], cmap='viridis')
axes[1, 0].set_title(r'$|\psi_1^{(0)\prime}|^2 = \frac{1}{\sqrt{2}} (\psi_{2,1} + \psi_{1,2})^2$')
axes[1, 0].set_xlabel('x')
axes[1, 0].set_ylabel('y')
plt.colorbar(im3, ax=axes[1, 0])

# Plot psi_2_prime
im4 = axes[1, 1].imshow(prob_2_prime, origin='lower', extent=[0, L, 0, L], cmap='viridis')
axes[1, 1].set_title(r'$|\psi_2^{(0)\prime}|^2 = \frac{1}{\sqrt{2}} (\psi_{2,1} - \psi_{1,2})^2$')
axes[1, 1].set_xlabel('x')
axes[1, 1].set_ylabel('y')
plt.colorbar(im4, ax=axes[1, 1])

plt.tight_layout()
plt.show()
```
