```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt

# Define a simple model
model = nn.Sequential(nn.Linear(10, 5), nn.ReLU(), nn.Linear(5, 1))
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
loss_fn = nn.MSELoss()

# Fake data
x = torch.randn(20, 10)
y = torch.ones(20, 1)
losses = []
predictions = []

# Train
for _ in range(100):
    pred = model(x)
    loss = loss_fn(pred, y)
    losses.append(loss.item()) #Store the loss value
    predictions.append(pred) #Store the predictions
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
print("Done!")

# Define a simple model
model = nn.Sequential(nn.Linear(10, 5), nn.ReLU(), nn.Linear(5, 1))
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
loss_fn = nn.MSELoss()

# Fake data
x = torch.randn(20, 10)
y = torch.ones(20, 1)

# Train
for _ in range(100):
    pred = model(x)
    loss = loss_fn(pred, y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
print("Done!")

plt.plot(losses)
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()

for i, pred in enumerate(predictions):
    plt.plot(pred.detach().numpy(), label=f"Epoch {i+1}")  # Detach from computation graph and convert to numpy
plt.title("Model Predictions")
plt.xlabel("Data Point")
plt.ylabel("Prediction")
plt.legend()
plt.show()
```
