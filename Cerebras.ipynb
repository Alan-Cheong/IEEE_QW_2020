```python
import os
from cerebras.cloud.sdk import Cerebras

client = Cerebras(
    api_key=""
)

completion = client.chat.completions.create(
    messages=[{"role":"user","content":"Why is fast inference important?"}],
    model="llama-3.3-70b",
    max_completion_tokens=1024,
    temperature=0.2,
    top_p=1,
    stream=False
)

print(completion.choices[0].message.content)
```

Once installed, it's best practice to store your API key securely using Colab's Secrets Manager. You can find it under the 'ðŸ”‘' icon in the left panel. Create a new secret and name it `CEREBRAS_API_KEY` (or any other name you prefer). Then, you can access it in your code like this:

```python
# Import necessary libraries
import os
from cerebras.cloud.sdk import Cerebras
from google.colab import userdata

# Retrieve the API key from Colab secrets
CEREBRAS_API_KEY = userdata.get('CEREBRAS_API_KEY')

# Initialize the Cerebras client with your API key
client = Cerebras(api_key=CEREBRAS_API_KEY)

print("Cerebras client initialized successfully!")
```

Now that the client is initialized, you can make calls to the Cerebras models. Here's an example of how to generate text using the `llama-3.3-70b` model:

```python
completion = client.chat.completions.create(
    messages=[{"role":"user","content":"Explain the concept of quantum entanglement in simple terms."}],
    model="llama-3.3-70b",
    max_completion_tokens=500,
    temperature=0.7,
    top_p=1,
    stream=False
)

print(completion.choices[0].message.content)
```
